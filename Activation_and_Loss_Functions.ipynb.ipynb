```python
import numpy as np
import matplotlib.pyplot as plt

def step(x):
    return np.where(x >= 0, 1, 0)

def step_derivative(x):
    return np.zeros_like(x)
    
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, step(x), label='Step Function')
plt.title('Step Function')
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(x, step_derivative(x), label='Derivative', color='red')
plt.title('Step Function Derivative')
plt.grid()
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sig = sigmoid(x)
    return sig * (1 - sig)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, sigmoid(x), label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(x, sigmoid_derivative(x), label='Derivative', color='red')
plt.title('Sigmoid Function Derivative')
plt.grid()
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, tanh(x), label='Tanh Function')
plt.title('Tanh Function')
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(x, tanh_derivative(x), label='Derivative', color='red')
plt.title('Tanh Function Derivative')
plt.grid()
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, relu(x), label='ReLU Function')
plt.title('ReLU Function')
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(x, relu_derivative(x), label='Derivative', color='red')
plt.title('ReLU Function Derivative')
plt.grid()
plt.show()
```

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, leaky_relu(x), label='Leaky ReLU Function')
plt.title('Leaky ReLU Function')
plt.grid()
plt.subplot(1, 2, 2)
plt.plot(x, leaky_relu_derivative(x), label='Derivative', color='red')
plt.title('Leaky ReLU Function Derivative')
plt.grid()
plt.show()
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

x_softmax = np.linspace(-2, 2, 100)
softmax_values = softmax(x_softmax)

plt.figure(figsize=(5, 5))
plt.plot(x_softmax, softmax_values, label='Softmax Function')
plt.title('Softmax Function')
plt.grid()
plt.show()
```

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

y_true = np.array([1, 0, 1, 0])
y_pred = np.array([0.8, 0.1, 0.9, 0.3])
mse = mean_squared_error(y_true, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15 
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

bce = binary_cross_entropy(y_true, y_pred)
print(f'Binary Cross-Entropy: {bce}')
```

```python
def categorical_cross_entropy(y_true, y_pred):
    epsilon = 1e-15 
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(np.sum(y_true * np.log(y_pred), axis=1))

y_true_categorical = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
y_pred_categorical = np.array([[0.7, 0.2, 0.1], [0.1, 0.8, 0.1], [0.2, 0.3, 0.5]])
cce = categorical_cross_entropy(y_true_categorical, y_pred_categorical)
print(f'Categorical Cross-Entropy: {cce}')
```
